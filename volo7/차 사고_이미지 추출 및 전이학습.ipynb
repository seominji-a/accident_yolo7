```python
img_list=glob.glob('/content/drive/MyDrive/data/accident/images/*.jpg')
print(len(img_list))
```

```python
## 데이터 train/test/vaild 로 나누기
from sklearn.model_selection import train_test_split
train_img_lst, valid_img_lst = train_test_split(img_list, test_size=0.4, random_state=111)
valid_img_lst, test_img_lst = train_test_split(valid_img_lst, test_size=0.2, random_state=222)

print(len(train_img_lst), len(valid_img_lst), len(test_img_lst))
```

```python
%pwd
```

```python
# 데이터 정보 저장하기
with open('/content/drive/MyDrive/data/accident/train.txt', 'w') as f:
    f.write('\n'.join(train_img_lst)+'\n')

with open('/content/drive/MyDrive/data/accident/valid.txt', 'w') as f:
    f.write('\n'.join(valid_img_lst)+'\n')

with open('/content/drive/MyDrive/data/accident/test.txt', 'w') as f:
    f.write('\n'.join(test_img_lst)+'\n')
```
